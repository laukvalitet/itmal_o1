```python
from matplotlib import pyplot
from sklearn import datasets


iris = datasets.load_iris()

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


#Basic train test split, no cross validation
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target)
knn = KNeighborsClassifier(n_neighbors=5)
model = knn.fit(X_train,y_train)
y_predicted = model.predict(X_test)
print(model.score(X_test,y_test))
k_range = range(1,113)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    model = knn.fit(X_train,y_train)
    y_predicted = model.predict(X_test)
    scores.append(model.score(X_test,y_test))
pyplot.bar(k_range,scores)
    
```

```python
%reset
from sklearn import datasets
from sklearn.model_selection import cross_val_score
from matplotlib import pyplot
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

iris = datasets.load_iris()

k_range = range(1,113)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores_for_this_k_val = cross_val_score(knn,iris.data,iris.target,cv=10)
    scores.append(scores_for_this_k_val.mean())
pyplot.bar(k_range,scores)
#print(scores)
print("BEST K VAL:", np.argmax(scores)+1) # +1 because range starts at 1
```
